## Q1: How to identify VIP & Preferred Customers?

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# convert 'amount_spent' column into numeric format
order['amount_spent'] = pd.to_numeric(order['amount_spent'], errors='coerce')

#Group by CustomerID and calculate the total amount spent by customers
customer_expenses = order.groupby('CustomerID')['amount_spent'].sum().dropna()  # Remove NaN values


#Determine the quantile thresholds using qcut
num_quantiles = 3  # Number of quantiles (including the minimum and maximum)
labels = pd.qcut(customer_expenses, q=num_quantiles, labels=["Regular", "Preferred", "VIP"])

# Add the labels to the customer_expenses Series
customer_expenses_labeled = customer_expenses.to_frame(name='customer_expenses')
customer_expenses_labeled['Customer Type'] = labels

# Display the labeled customers
customer_expenses_labeled

# Merge customer_labels DataFrame with customer_expenses DataFrame
order = pd.merge(order, customer_expenses_labeled, left_on='CustomerID', right_index=True)

# Display the DataFrame with Customer Type information
order
```

## Q2: How to identify which country has the most VIP Customers?

```python
# Filter the DataFrame to include only VIP customers
vip_customers = order[order['Customer Type'] == 'VIP']

# Group DataFrame by country and count the number of VIP customers in each country
vip_cust_country = vip_customers.groupby('Country').value_counts()

# Find the country with the most VIP customers
most_vip_customers = vip_cust_country.idxmax()
most_vip_customers

# Getting Product space too large to allocate arrays!
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# Filter the DataFrame to include only VIP and both preferred customers
vip_pref_customers = order[order['Customer Type'] == 'VIP']

# Group DataFrame by country and count the number of VIP customers in each country
vip_pref_cust_country = vip_pref_customers.groupby('Country').value_counts()

# Find the country with the most VIP customers
most_vip_pref_customers = vip_pref_cust_country.idxmax()
most_vip_pref_customers
```
